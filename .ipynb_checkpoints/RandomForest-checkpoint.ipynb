```python
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
health = sns.load_dataset('healthexp')
```

```python
health.head(10)
```

```python
health = pd.get_dummies(health)
```

```python
health.head(10)
```

```python
x = health.drop(['Life_Expectancy'], axis = 1)
```

```python
x.head(10)
```

```python
y = health['Life_Expectancy']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 19)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state=13)
```

```python
rfr.fit(x_train,y_train)
```

```python
y_pred = rfr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
mean_squared_error(y_pred,y_test)
```

```python
r2_score(y_pred,y_test)
```

```python
param_grid = {
    'n_estimators': [100,200,300],
    'max_depth':[10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rfr_cv = GridSearchCV(estimator = rfr, param_grid = param_grid, cv = 3, scoring = 'neg_mean_squared_error', n_jobs = -1)
```

```python
rfr_cv.fit(x_train,y_train)
```

```python
y_pred = rfr_cv.predict(x_test)
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
mean_squared_error(y_pred,y_test)
```

```python
r2_score(y_pred,y_test)
```

```python
importances = rfr.feature_importances_

# Create a DataFrame to sort and visualize feature importances
feature_importance_df = pd.DataFrame({
    'Feature': x.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(feature_importance_df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature Importance')
plt.title('Feature Importance in Random Forest')
plt.show()
```
